```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
df = pd.read_csv("/content/train.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['Survived'].value_counts()
df['Sex'].value_counts()
df['Pclass'].value_counts()
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
df.drop('Cabin', axis=1, inplace=True)
```

```python
sns.countplot(x='Survived', data=df)
plt.title("Survival Count")
plt.show()
```

```python
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Survival by Gender")
plt.show()
```

```python
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title("Survival by Class")
plt.show()
```

```python
plt.hist(df['Age'], bins=20)
plt.title("Age Distribution")
plt.show()
```

```python
plt.hist(df['Fare'], bins=20)
plt.title("Fare Distribution")
plt.show()
```

```python
sns.boxplot(x='Survived', y='Fare', data=df)
plt.show()
```

```python
# Convert Sex to numeric
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})

# Convert Embarked to numeric
df['Embarked'] = df['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.show()
```

```python
sns.pairplot(df[['Survived','Age','Fare','Pclass']], hue='Survived')
plt.show()
```

```python
print("EDA Summary:")
print("1. Females had higher survival rate.")
print("2. 1st class passengers survived more.")
print("3. Higher fare increased survival chance.")
print("4. Most passengers were aged 20-40.")
```

```python
# Create Age Groups
df['AgeGroup'] = pd.cut(df['Age'], bins=[0,12,18,35,60,80])

sns.barplot(x='AgeGroup', y='Survived', data=df)
plt.title("Survival Rate by Age Group")
plt.xticks(rotation=45)
plt.show()
```

```python
df['FareGroup'] = pd.qcut(df['Fare'], 4)

sns.barplot(x='FareGroup', y='Survived', data=df)
plt.title("Survival Rate by Fare Group")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.violinplot(x='Survived', y='Age', data=df)
plt.title("Age Distribution by Survival")
plt.show()
```
